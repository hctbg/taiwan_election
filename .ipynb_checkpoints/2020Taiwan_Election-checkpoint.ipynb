# How many people vote in each candidate

```python
#how many people voted validly in Taiwan 
total_votes = results.vote.sum()
```

```python
print(f'There were {total_votes:,} cast in Taiwan Presidential Election.')
```

## Who won the election

```python
cadidate_totals = results.groupby('name').vote.sum().reset_index().sort_values(
    'vote', ascending=False
)
cadidate_totals
cadidate_totals['name'].replace(r"蔡英文賴清德", "Ing-Wen, Tsai", inplace = True)
cadidate_totals['name'].replace(r"韓國瑜張善政", "Kuo-yu, Han", inplace = True)
cadidate_totals['name'].replace(r"宋楚瑜余湘", "Chu-yu, Soong ", inplace = True)
cadidate_totals
```

```python
cadidate_totals['VoteShare'] = round(
    (cadidate_totals['vote'] / total_votes) * 100
    ,2
)
```

```python
cadidate_totals
```

# Local_Winner: Winner for each place

```python
# results = results.sort_values('per', ascending=False)
winners = results.groupby("place",sort = False).per.idxmax()
# winners = results.drop_duplicates('place', keep='first')
df = results.iloc[winners,:]
df = df.reset_index()
del df['index']
del df['party']
del df['name']
df["number"].replace(3,"Democratic Progressive Party",  inplace = True)
df["number"].replace(2,"Kuomintang",  inplace = True)
df["order"].replace(3, "Ing-Wen, Tsai", inplace = True)
df["order"].replace(2, "Kuo-yu, Han", inplace = True)
df.columns = ['place','candidate_name','gender','vote','percentage','party','local_total']
df['gender'].replace(r"女男", "F/M", inplace = True)
df['gender'].replace(r"男男", "M/M", inplace = True)
df
```

## Access county-level boundary data

```python
Taiwan_map = gpd.read_file('/Users/xiangyuchen/Taiwan-election/taiwan.geojson')
```

```python
gdf = gpd.GeoDataFrame.from_features(Taiwan_map)
```

```python
gdf.loc[0,'name_traditional_chinese'] = ['臺北市']
gdf.loc[2,'name_traditional_chinese'] = ['臺中市']
gdf.loc[3,'name_traditional_chinese'] = ['臺南市']
gdf.loc[21,'name_traditional_chinese'] = ['臺東縣']
gdf.loc[1,['name', 'name_traditional_chinese']] = ['Taoyuan City', '桃園市']
del gdf['updated_at']
del gdf['created_at']
gdf
```
